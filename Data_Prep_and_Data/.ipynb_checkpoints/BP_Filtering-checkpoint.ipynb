```python
import scipy.io
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import savemat
from mne.filter import filter_data
import seaborn as sns
```

```python
raw_eeg = r"D:\SEED-IV\eeg_raw_data\3\1_20161126.mat"
raw_data = scipy.io.loadmat(raw_eeg)
print(raw_data.keys())
```

```python
print(f"Original data type: {raw_data['cz_eeg1'].dtype}")
print(f"Original data size (MB): {raw_data['cz_eeg1'].nbytes / 1e6:.2f}")
```

```python
filter_eeg_dict = {}
for key in raw_data:
    if key.startswith('cz_eeg'):
        filtered = filter_data(raw_data[key], sfreq = 200, l_freq = 0.5, h_freq = 47.0, verbose = True)
        filter_eeg_dict[key] = filtered
```

```python

savemat(r'C:\Users\Noman\Desktop\Github\CTL_Scratch\Data\3\1_BP_filtered_3.mat', filter_eeg_dict)
```
